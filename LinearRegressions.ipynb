```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset.csv')
df.head(5)
```

```python
describe = df.describe()
info = df.info()

print(describe)
print(info)
```

```python

x = df['Weight'].values.reshape(-1,1)

y = df['Height'].values
```

```python
b_train, b_test, y_train, y_test = train_test_split( x, y, train_size=0.8, test_size=0.2 ,random_state=22)
```

```python
model = LinearRegression()
model.fit(b_train, y_train)
```

```python
y_predict = model.predict(b_test)
```

```python
plt.scatter(b_train, y_train, color='green')
plt.plot(b_test, y_predict, color='red')
plt.ylabel('Salary')
plt.xlabel('Years Of Expreince')
plt.title('Training Data vs Model')
plt.show()
```

```python
x = df['Width'].values.reshape(-1,1)

y = df['Height'].values

b_train, b_test, y_train, y_test = train_test_split( x, y, train_size=0.8, test_size=0.2 ,random_state=14)

model = LinearRegression()
model.fit(b_train, y_train)

y_predict = model.predict(b_test)
plt.scatter(b_train, y_train, color='green')
plt.plot(b_test, y_predict, color='red')
plt.xlabel('Width of the fish')
plt.ylabel('Height of the fish')
plt.title('Training Data vs Model')
plt.show()
```
